## Define special measurement and preparation basis for QB native gates

```python
# -*- coding: utf-8 -*-

# This code is part of Qiskit.
#
# (C) Copyright IBM 2019.
#
# This code is licensed under the Apache License, Version 2.0. You may
# obtain a copy of this license in the LICENSE.txt file in the root directory
# of this source tree or at http://www.apache.org/licenses/LICENSE-2.0.
#
# Any modifications or derivative works of this code must retain this
# copyright notice, and modified files need to carry a notice indicating
# that they have been altered from the originals.

"""Pauli tomography preparation and measurement basis
"""

# Needed for functions
import numpy as np

# Import QISKit classes
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.ignis.verification.tomography.basis import TomographyBasis


###########################################################################
# Built-in circuit functions
###########################################################################

def qb_pauli_measurement_circuit(
        op: str,
        qubit: QuantumRegister,
        clbit: ClassicalRegister
) -> QuantumCircuit:
    """Return a qubit Pauli operator measurement circuit.

    Params:
        op: Pauli operator 'X', 'Y', 'Z'.
        qubit: qubit to be measured.
        clbit: clbit for measurement outcome.

    Returns:
        The measurement circuit for the given Pauli.
    """

    circ = QuantumCircuit([qubit, clbit])
    if op == 'X':
        circ.ry(-np.pi/2,qubit)
        circ.measure(qubit, clbit)
    if op == 'Y':
        circ.rx(np.pi/2,qubit)
        circ.measure(qubit, clbit)
    if op == 'Z':
        circ.measure(qubit, clbit)
    return circ


def qb_pauli_preparation_circuit(
        op: str,
        qubit: QuantumRegister
) -> QuantumCircuit:
    """Return a qubit Pauli eigenstate preparation circuit.

    This circuit assumes the qubit is initialized
    in the :math:`Zp` eigenstate :math:`[1, 0]`.

    Params:
        op: Pauli eigenstate 'Zp', 'Zm', 'Xp', 'Xm', 'Yp', or 'Ym'.
        qubit: qubit to be prepared.

    Returns:
        The preparation circuit for the given Pauli eigenstate.
    """

    circ = QuantumCircuit([qubit])
    if op == 'Xp':
        circ.ry(np.pi/2,qubit)
    if op == 'Xm':
        circ.ry(-np.pi/2,qubit)
    if op == 'Yp':
        circ.rx(-np.pi/2,qubit)
    if op == 'Ym':
        circ.ry(np.pi/2,qubit)
    if op == 'Zm':
        circ.ry(np.pi,qubit)
    return circ


###########################################################################
# Matrix functions for built-in bases
###########################################################################

def qb_pauli_preparation_matrix(label: str) -> np.array:
    r"""Return the matrix corresponding to a Pauli eigenstate preparation.

    Args:
        label: single-qubit Pauli eigenstate operator label.

    Returns:
        A Numpy array for the Pauli eigenstate. Allowed inputs
            and corresponding returned matrices are:

            'Xp' : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & 1\\1 & 1\end{array}\right)`

            'Xm' : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & -1\\-1 & 1\end{array}\right)`

            'Yp' : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & -i\\i & 1\end{array}\right)`

            'Ym' : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & i\\-i & 1\end{array}\right)`

            'Zp' : :math:`\left(\begin{array}
            {cc}1 & 0\\0 & 0\end{array}\right)`

            'Zm' : :math:`\left(\begin{array}
            {cc}0 & 0\\0 & 1\end{array}\right)`
    """
    res = np.array([])
    # Return matrix for allowed label
    if label == 'Xp':
        res = np.array([[0.5, 0.5], [0.5, 0.5]], dtype=complex)
    if label == 'Xm':
        res = np.array([[0.5, -0.5], [-0.5, 0.5]], dtype=complex)
    if label == 'Yp':
        res = np.array([[0.5, -0.5j], [0.5j, 0.5]], dtype=complex)
    if label == 'Ym':
        res = np.array([[0.5, 0.5j], [-0.5j, 0.5]], dtype=complex)
    if label == 'Zp':
        res = np.array([[1, 0], [0, 0]], dtype=complex)
    if label == 'Zm':
        res = np.array([[0, 0], [0, 1]], dtype=complex)
    return res


def qb_pauli_measurement_matrix(label: str, outcome: int) -> np.array:
    r"""Return the matrix corresponding to a Pauli measurement outcome.

    Args:
        label: single-qubit Pauli measurement operator label.
        outcome: measurement outcome.

    Returns:
        A Numpy array for measurement outcome operator.
            Allowed inputs and corresponding returned matrices are:

            'X', 0 : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & 1\\1 & 1\end{array}\right)`

            'X', 1 : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & -1\\-1 & 1\end{array}\right)`

            'Y', 0 : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & -i\\i & 1\end{array}\right)`

            'Y', 1 : :math:`\frac{1}{2}
            \left(\begin{array}{cc}1 & i\\-i & 1\end{array}\right)`

            'Z', 0 : :math:`\left(\begin{array}
            {cc}1 & 0\\0 & 0\end{array}\right)`

            'Z', 1 : :math:`\left(\begin{array}
            {cc}0 & 0\\0 & 1\end{array}\right)`
    """
    res = np.array([])
    # Return matrix
    if label == 'X':
        if outcome in ['0', 0]:
            res = pauli_preparation_matrix('Xp')
        if outcome in ['1', 1]:
            res = pauli_preparation_matrix('Xm')
    if label == 'Y':
        if outcome in ['0', 0]:
            res = pauli_preparation_matrix('Yp')
        if outcome in ['1', 1]:
            res = pauli_preparation_matrix('Ym')
    if label == 'Z':
        if outcome in ['0', 0]:
            res = pauli_preparation_matrix('Zp')
        if outcome in ['1', 1]:
            res = pauli_preparation_matrix('Zm')
    return res


###########################################################################
# PauliBasis Object
###########################################################################

QBPauliBasis = TomographyBasis('QB_Pauli',  # pylint: disable=invalid-name
                             measurement=(('X', 'Y', 'Z'),
                                          qb_pauli_measurement_circuit,
                                          qb_pauli_measurement_matrix),
                             preparation=(('Xp', 'Yp', 'Zp', 'Zm'),
                                          qb_pauli_preparation_circuit,
                                          qb_pauli_preparation_matrix))
```

## Utilities

```python
def get_qbqasm_string(qiskit_circuit):
    '''
    Input: Qiskit circuit object
    Output: QBCIRCUIT string
    '''
    import re
    return '__qpu__ void QBCIRCUIT(qreg q) {\n' + re.sub(r"\nqreg [A-Za-z]+[_\dA-za-z]*\[\d+\];", "", qiskit_circuit.qasm()) + '}'

def get_qiskit_experiment_result_data(qbos_result, num_qubits):
    '''
    Input: dict of counts for a qbos experiment
    Output: qiskit.result.models.ExperimentResultData containing counts keyed with hex code
    
    The qbos enumeration from out_raw to out_count is reversed compared to Qiskit
    '''
    hex_keyed_data = dict()
    shots = 0
    for num in qbos_result.keys():
        hex_keyed_data[hex(int(bin(num)[2:].zfill(num_qubits)[::-1],2))] = qbos_result[num]
        shots = shots + qbos_result[num]
    experiment_result_data = qiskit.result.models.ExperimentResultData(counts=hex_keyed_data)
    return experiment_result_data, shots
    

def get_experiment_result(qbos_core):
    '''
    Input: qbos.core
    Output: qiskit.result.result.Result
    '''
    experiment_result_data = []
    qbos_counts = qbos_core.out_count
    #print(qbos_counts)
    i = 0
    for row in qbos_counts:
        j = 0
        for result in row:
            num_qubits = qbos_core.qn[i][j]
            name = qbos_core.name_p[i][j]
            data, shots=get_qiskit_experiment_result_data(result, num_qubits)
            experiment_result_data.append(qiskit.result.models.ExperimentResult(data=data, shots=shots, success=True, header=qiskit.qobj.common.QobjHeader(name = name, memory_slots=num_qubits)))
            j+=1
        i+=1    
    experiment_result = qiskit.result.result.Result(backend_name=qbos_core.acc[0][0], backend_version=None, qobj_id=None, job_id=None, success=True,results=experiment_result_data)
    return experiment_result

def qiskit_calibration_circuits(num_qubits, qubit_list, experiment_name = 'readout_', quantum_register='q', classical_register='q_c'):
    from qiskit.ignis.mitigation.measurement import complete_meas_cal
    qr = qiskit.QuantumRegister(num_qubits, quantum_register)
    cr = qiskit.ClassicalRegister(num_qubits, classical_register)
    meas_calibs, state_labels = complete_meas_cal(qubit_list=qubit_list, qr=qr, cr=cr, circlabel=experiment_name)
    return meas_calibs, state_labels

def qiskit_state_tomography_circuits(circuit, quantum_register = None, meas_labels='Pauli', meas_basis='Pauli'):
    if quantum_register is None:
        quantum_register = circuit.qregs
    from qiskit.ignis.verification.tomography import state_tomography_circuits
    circuits = state_tomography_circuits(circuit, quantum_register, meas_labels=meas_labels, meas_basis=meas_basis)    
    return circuits

def qiskit_process_tomography_circuits(circuit, quantum_register = None, prepared_qubits=None, meas_labels='Pauli', meas_basis='Pauli', prep_labels='Pauli', prep_basis='Pauli'):
    if quantum_register is None:
        quantum_register = circuit.qregs
    from qiskit.ignis.verification.tomography import process_tomography_circuits
    circuits = process_tomography_circuits(circuit, quantum_register, prepared_qubits=prepared_qubits, meas_labels=meas_labels, meas_basis=meas_basis, prep_labels=prep_labels, prep_basis=prep_basis)
    return circuits

def qiskit_gate_set_tomography_circuits(quantum_register = None, gateset_basis='default'):
    from qiskit.ignis.verification.tomography import gateset_tomography_circuits
    circuits = process_tomography_circuits(quantum_register, gateset_basis)
    return circuits

def get_qbos_core_from_qiskit_circuits(circuits, backend, no_shots):
    tqb = qb.core()
    tqb.qb12()
    tqb.sn = no_shots
    tqb.acc = backend
    tqb.instring.clear()
    tqb.noise.clear()
    tqb.qn.clear()
    tqb.name_p.clear()
    for circuit in circuits:
        tqb.qn.append(qb.N([circuit.num_qubits]))
        tqb.noise.append(qb.Bool([True]))
        tqb.instring.append(qb.String([get_qbqasm_string(circuit)])) # TODO: use Thien's QASM method
        tqb.name_p.append(qb.String([circuit.name]))   
    return tqb

def get_measurement_fitter(num_qubits, qubit_list, experiment_name, no_shots, backend, quantum_register='q', classical_register='q_c'):
    meas_calibs, state_labels =  qiskit_calibration_circuits(num_qubits, qubit_list, experiment_name, quantum_register, classical_register)
    # for circ in meas_calibs:
    #     print(circ) # display circuits for debugging
    tqb = get_qbos_core_from_qiskit_circuits(meas_calibs, backend, no_shots)
    tqb.run()
    qiskit_result = get_experiment_result(tqb)
    # Calculate the calibration matrix with the noise model
    from qiskit.ignis.mitigation.measurement import CompleteMeasFitter
    meas_fitter = CompleteMeasFitter(qiskit_result, state_labels, qubit_list=qubit_list, circlabel=experiment_name)
    return meas_fitter

def get_state_tomography_fitter(circuit, backend, no_shots, error_mitigation=False, calibration_shots = -1, mitigation_method = 'least_squares', meas_labels='Pauli', meas_basis='Pauli'):
    from qiskit.ignis.verification import state_tomography_circuits, StateTomographyFitter
    qst = state_tomography_circuits(circuit, circuit.qregs, meas_labels, meas_basis)
    # for circ in qst:
    #     print(circ) # display circuits for debugging
    tqb = get_qbos_core_from_qiskit_circuits(qst, backend, no_shots)
    tqb.run()
    qiskit_result = get_experiment_result(tqb)
    if error_mitigation:
        if calibration_shots < 0:
            calibration_shots = no_shots
        meas_fitter = get_measurement_fitter(num_qubits, qubit_list, 'readout_', calibration_shots, backend)
        qiskit_result = meas_fitter.filter.apply(qiskit_result, method=mitigation_method)
    tomo_fitter = StateTomographyFitter(qiskit_result, qst)
    return tomo_fitter

def get_process_tomography_fitter(circuit, backend, no_shots, error_mitigation=False, calibration_shots = -1, mitigation_method = 'least_squares', meas_labels='Pauli', meas_basis='Pauli', prep_labels='Pauli', prep_basis='Pauli'):
    from qiskit.ignis.verification.tomography import process_tomography_circuits, ProcessTomographyFitter
    qpt = process_tomography_circuits(circuit, measured_qubits=circuit.qregs, prepared_qubits=circuit.qregs, meas_labels=meas_labels, meas_basis=meas_basis, prep_labels=prep_labels, prep_basis=prep_basis)
    #for circ in qpt:
    #    print(circ) # display circuits for debugging
    tqb = get_qbos_core_from_qiskit_circuits(qpt, backend, no_shots)
    tqb.run()
    qiskit_result = get_experiment_result(tqb)
    if error_mitigation:
        if calibration_shots < 0:
            calibration_shots = no_shots
        meas_fitter = get_measurement_fitter(num_qubits, qubit_list, 'readout_', calibration_shots, backend)
        qiskit_result = meas_fitter.filter.apply(qiskit_result, method=mitigation_method)
    tomo_fitter = ProcessTomographyFitter(qiskit_result, qpt)
    return tomo_fitter
```

# State Tomography

```python
import qbos as qb
import qiskit
#print(qiskit.__qiskit_version__)

num_qubits = 1
no_shots = 700 # Note number of shots is calculated from actual count data returned from the backend
backend = 'aer'
fitter_method = 'lstsq'
error_mitigation = True # Note that the calibration circuits are run after the state tomography circuits, can be easily changed in get_state_tomography_fitter
calibration_shots = 700 # Note number of shots is calculated from actual count data returned from the backend
mitigation_method = 'least_squares'
meas_labels = QBPauliBasis.measurement_labels
meas_basis = QBPauliBasis
qubit_list = [0]
qr_process = qiskit.QuantumRegister(num_qubits, 'q')
process = qiskit.QuantumCircuit(qr_process)
process.ry(np.pi/2,qr_process[0])
# process.cx(qr_process[0], qr_process[1])


tomo_fitter = get_state_tomography_fitter(process, backend, no_shots, error_mitigation, calibration_shots, mitigation_method, meas_labels, meas_basis)
rho_fit = tomo_fitter.fit(method=fitter_method)
F_circuit = qiskit.qiskit.quantum_info.state_fidelity(rho_fit, qiskit.qiskit.quantum_info.Statevector.from_instruction(process))
print('State fidelity: F = {:.5f}'.format(F_circuit))
from qiskit.visualization import plot_state_city
plot_state_city(rho_fit, color=['midnightblue', 'midnightblue'])
```

# Process tomography

```python
import qbos as qb
import qiskit
#print(qiskit.__qiskit_version__)

num_qubits = 1
no_shots = 700 # Note number of shots is calculated from actual count data returned from the backend
backend = 'aer'
fitter_method = 'lstsq'
error_mitigation = True # Note that the calibration circuits are run after the state tomography circuits, can be easily changed in get_state_tomography_fitter
calibration_shots = 700 # Note number of shots is calculated from actual count data returned from the backend
mitigation_method = 'least_squares'
meas_labels = QBPauliBasis.measurement_labels
meas_basis = QBPauliBasis
prep_labels = QBPauliBasis.preparation_labels
prep_basis = QBPauliBasis
qubit_list = [0]
qr_process = qiskit.QuantumRegister(num_qubits, 'q')
process = qiskit.QuantumCircuit(qr_process)
process.ry(np.pi/2,qr_process[0])
#process.cx(qr_process[0], qr_process[1])


tomo_fitter = get_process_tomography_fitter(process, backend, no_shots, error_mitigation, calibration_shots, mitigation_method, meas_labels, meas_basis, prep_labels, prep_basis)
choi_fit_lstsq = tomo_fitter.fit(method=fitter_method)
F_circuit = qiskit.qiskit.quantum_info.average_gate_fidelity(choi_fit_lstsq, target=qiskit.quantum_info.Operator(process))
print('Process fidelity: F = {:.5f}'.format(F_circuit))
```
