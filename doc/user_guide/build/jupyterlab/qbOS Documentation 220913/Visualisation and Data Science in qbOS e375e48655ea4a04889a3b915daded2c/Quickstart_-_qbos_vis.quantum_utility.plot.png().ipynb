## Quickstart guide for `qbos_vis.quantum_utility.plot.png()`

This function produces a Quantum Utility visualisation and saves the resulting plot in a .png format image file.

### Import statement
```
import qbos_vis
```
### Inputs
#### - Required inputs
```
qn : number of qubits [list of integer values]
exec_ms : quantum execution time, in ms [list of real values]
classical_exec_ms : classical algorithm execution time, in ms [list of real values]
fidelity : quantum algorithm accuracy [list of real values between 0 and 1]
classical_fidelity : classical algorithm accuracy [list of real values betweeen 0 and 1]
fidelity_std : quantum algorithm precision [list of real values]
classical_fidelity_std : classical algorithm precision [list of real values]
```
#### - Optional inputs
```
in_dpi : dots-per-inch resolution of the output image [default: 300]
in_title : text to display at the top of the output image [default: "Set by providing: in_title"]
in_data_name : text to display alongside the data points [default: "Set by providing: in_data_name"]
```
### Calling the function and what it does to the inputs
```
qbos_vis.quantum_utility.plot.png(qn, exec_ms, classical_exec_ms, fidelity, classical_fidelity, fidelity_std, classical_fidelity_std, in_title="your title", in_data_name="your dataset name")
```
* The inputs are visualised as a scatter plot
* The x-axis maps to 'Relative Speed' := classical_exec_ms/exec_ms
* The y-axis maps to 'Relative Accuracy' := fidelity/classical_fidelity
* The shade of the data points maps to 'Relative Precision' := fidelity_std/classical_fidelity_std (note: we equate 'precision' to the standard deviation of fidelity)

### Outputs

An image file with filename: `<in_data_name>_<in_title>_quantum_utility.png`
