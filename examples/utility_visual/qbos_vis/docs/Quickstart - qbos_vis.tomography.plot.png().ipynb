## Quickstart guide for `qbos_vis.tomography.plot.png()`

This function produces a Tomography visualisation and saves the resulting plot in a .png format image file.

### Import statement
```
import qbos_vis
```
### Inputs
#### - Required inputs
```
rho : density matrix [matrix of complex values]
rho_std : standard deviation of density matrix [list of real values]
```

#### - Optional inputs
```
output_filename : name of the file in which to save the image (a .png extension is automatically added) [default: "Set by providing: output_filename"]
```

### Calling the function and what it does to the inputs
```
This code accepts a density matrix (of complex values) and standard deviation of density matrix. It computes the magnitude of each element of rho and then uses the result to create a 3D bar chart and shows error bars calculated from +/-1 standard deviation. 
```

### Outputs

A tomography 3D bar chart for every quantum state, saved in a file named: `<output_filename>.png`


### Examples
####  Example 1: Tomography

````python
import qbos_vis
# Dataset
rho = [9.8987e-01+0.j, 3.3000e-04-0.00188j, 3.3000e-04+0.00188j, 1.0130e-02+0.j]
rho_std =[0.0016862013521522288, 0.021686094623052828, 0.021686094623052828, 0.0016862013521522288]
    
qbos_vis.tomography.plot.png( rho, rho_std)

````
